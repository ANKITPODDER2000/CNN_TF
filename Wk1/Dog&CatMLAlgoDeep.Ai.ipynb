```python
import numpy as np
```

```python
def split_test_data(random_state=42,test_size=0.1):
    
    np.random.seed(random_state)
    size1 = int(len(os.listdir(train_dog)) * test_size)
    size2 = int(len(os.listdir(train_cat)) * test_size)
    
    filedog = np.random.permutation(os.listdir(train_dog))[:size1]
    filecat = np.random.permutation(os.listdir(train_cat))[:size2]
    
    for dogs in filedog:
        shutil.move(train_dog+"/"+dogs,test_dog)
        
    for cats in filecat:
        shutil.move(train_cat+"/"+cats,test_cat)
```
